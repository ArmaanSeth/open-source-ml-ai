#### 1. Reading an image using imread() [imread(flag,path)]

```python
img = cv2.imread("python_logo.png")
cv2_imshow(img)
```

#### Imread flag attribute options
imread() has three options for the 'flag' attribute.

```python
#Load a color image - cv2.IMREAD_COLOR/1
#This is also the defauly flag
print("FLAG: cv2.IMREAD_COLOR/-1")
cv2_imshow(cv2.imread("python_logo.png",cv2.IMREAD_COLOR))
cv2_imshow(cv2.imread("python_logo.png",1))
```

```python
#Each pixel in a COLOR image is a Blue-Green-Red(BGR) triple.
#The range of each value in the triple is 0-255.
print(cv2.imread("python_logo.png",1)[0][0])
```

```python
#Load a grayscale image - cv2.IMREAD_GRAYSCALE/0

print("FLAG: cv2.IMREAD_GRAYSCALE/0")
cv2_imshow(cv2.imread("python_logo.png",cv2.IMREAD_GRAYSCALE))
cv2_imshow(cv2.imread("python_logo.png",0))
```

```python
#Each pixel in a GRAYSCALE image is a single value.
#Its range is 0-255.
print(cv2.imread("python_logo.png",0)[0][0])
```

```python
#Load a image unchanged - cv2.IMREAD_UNCHANGED/-1
#This is useful for reading images that have more than three channels
#For instance, some images have an fourth channel, alpha, which specifies the opacity of each color.
print("FLAG: cv2.IMREAD_UNCHANGED/-1")
cv2_imshow(cv2.imread("rgba_image.png",cv2.IMREAD_UNCHANGED))
cv2_imshow(cv2.imread("rgba_image.png",-1))
```

```python
#Each pixel has four values.
print(cv2.imread("rgba_image.png",-1)[0][0])
```

```python
print("IMREAD_COLOR shape",cv2.imread("rgba_image.png",1).shape)
print("IMREAD_GRAYSCALE shape",cv2.imread("rgba_image.png",0).shape)
print("IMREAD_UNCHANGED shape",cv2.imread("rgba_image.png",-1).shape)
```
